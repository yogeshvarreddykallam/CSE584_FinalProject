```python
from transformers import pipeline
import pandas as pd

# Load the questions from the Excel file
file_path = "only_question.xlsx"  # Ensure the file is in the same directory or provide the full path
questions_df = pd.read_excel(file_path)

# Validate that the file has a "questions" column
if "questions" not in questions_df.columns:
    raise ValueError("The Excel file must contain a 'questions' column.")

# Limit the DataFrame to the first 3 questions
# questions_df = questions_df.head(300)

# Load Hugging Face models as pipelines
model_pipelines = {
    "distilgpt2": pipeline("text-generation", model="distilgpt2"),
    "flan-t5": pipeline("text2text-generation", model="google/flan-t5-small"),
    "gpt-neo": pipeline("text-generation", model="EleutherAI/gpt-neo-1.3B")
}

# Define a function to get answers from a specific model
def get_answer(question, model_name):
    try:
        model = model_pipelines[model_name]
        # Generate an answer using the pipeline
        output = model(question, max_length=50, num_return_sequences=1)
        return output[0]["generated_text"]
    except Exception as e:
        print(e)
        return f"Error: {str(e)}"

# Apply each model to the questions and store answers in separate columns
for model_name in model_pipelines.keys():
    questions_df[f'answers_{model_name}'] = questions_df['questions'].apply(
        lambda q: get_answer(q, model_name)
    )

# Save the updated DataFrame with answers to a new Excel file
output_file_path = "questions_with_answers_huggingface.xlsx"
questions_df.to_excel(output_file_path, index=False)

print(f"Answers for the first 3 questions saved successfully to {output_file_path}")
```
